Спроецируем слова на плоскость через проекцию, сохраняющую как  можно больше расстояний

```python
tsne = TSNE(n_components=2,metric='cosine', n_iter=2500)
```

```python
selected_words = ["человек", "мужчина", "женщина",
                  "медсестра", "врач", "доктор",
                  "кошка", "кот", "жизнь", "смерть",
                  "грусть", "веселье", "хороший", "плохой"]
```

```python
selected_word_vectors = np.array([model[word] for word in selected_words])
```

```python
tsne.fit(selected_word_vectors)
```

```python
print(tsne.embedding_.shape)
```

```python
fig, ax = plt.subplots()
ax.scatter(tsne.embedding_[:,0], tsne.embedding_[:,1])
for word, point in zip(selected_words, tsne.embedding_):
    x,y = point
    ax.text(x,y,word)
ax.grid()
plt.show()
```

```python
tsne3d = TSNE(n_components=3,metric='cosine', n_iter=2500)
```

```python
tsne3d.fit(selected_word_vectors)
```

```python
fig = plt.figure()
ax = fig.add_subplot("111", projection="3d")
ax.scatter(tsne3d.embedding_[:,0], tsne3d.embedding_[:,1], tsne3d.embedding_[:,2])
for word, point in zip(selected_words, tsne3d.embedding_):
    x,y,z = point
    ax.text(x,y,z,word)
plt.show()
```

```python
pca = PCA(n_components=3)
```

```python
points = pca.fit_transform(selected_word_vectors)
```

```python
fig, ax = plt.subplots()
ax.scatter(points[:,0], points[:,1])
for word, point in zip(selected_words, points):
    x,y = point[:2]
    ax.text(x,y,word)
ax.grid()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot("111", projection="3d")
ax.scatter(points[:,0], points[:,1], points[:,2])
for word, point in zip(selected_words, points):
    x,y,z = point
    ax.text(x,y,z,word)
plt.show()
```
